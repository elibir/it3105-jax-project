```python
import sys
sys.path.append("../src/")
from sindres_wack_PID import *
import pandas as pd
```

```python
system = Consys(learning_rate=0.001, num_epochs=100, num_timesteps=100)
mse_history, pid_history = system.simulate_epochs()

# Convert JAX array to NumPy array before printing
mse_history_np = [mse.item() for mse in mse_history]
pid_history_np = [list(pid) for pid in pid_history]
```

```python
mse_dataframe = pd.DataFrame(mse_history_np)
pid_dataframe = pd.DataFrame(pid_history_np, columns=['kp','ki','kd'])

mse_dataframe.plot(figsize=(15,5))
pid_dataframe.plot(y=['kp','ki','kd'], figsize=(15,5))
```
